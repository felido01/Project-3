```python
# Import all Libraries that are necessary for data processing and modelling
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
card= pd.read_csv("creditcard.csv")
```

```python
card.head()
```

```python
card.shape
```

```python
card.dtypes
```

```python
#checking the balance of the dataset
fraud_count = card["Class"].value_counts()
fraud_rate = 100*fraud_count/card.shape[0]
fraud_data=pd.concat([fraud_count,fraud_rate], axis =1).reset_index()
fraud_data.columns =['Class', 'Count', 'Percentage']

fraud_data
```

```python
#handling imbalance 
card_fraud = card[card['Class']==1]
card_not_fraud = card[card['Class']==0]
card_not_fraud_sampled = card_not_fraud.sample(card_fraud.shape[0],replace=False, random_state=101)

card_balanced= pd.concat([card_not_fraud_sampled, card_fraud], axis= 0).sample(frac=1, replace= False, random_state=101).reset_index().drop('index',axis=1)
card_balanced
```

```python
# checking the balanced data
fraud_count= card_balanced["Class"].value_counts()
fraud_rate=100*fraud_count/card_balanced.shape[0]
fraud_data= pd.concat([fraud_count, fraud_rate], axis =1).reset_index()
fraud_data.columns =['Class','Count','Percentage']
fraud_data
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(card_balanced.drop('Class', axis=1),card_balanced['Class'], test_size =0.2, random_state=101)
print (f'''X_train:{X_train.shape}
X_test:{X_test.shape}
y_train:{y_train.shape}
y_test:{y_test.shape}''')
```

```python
# logistic model pipe
from sklearn.preprocessing import StandardScaler
randomForestModel = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', RandomForestClassifier())
])
randomForestModel.fit(X_train, y_train)
```

```python
#prediction
y_pred_logis = randomForestModel.predict(X_test)
y_pred_logis
```

```python
#classification report
cr = classification_report(y_test, y_pred_logis)
print(cr)
```

```python
#saving model
with open('./model.pkl', 'wb') as fp:
    pickle.dump(randomForestModel, fp)
```
